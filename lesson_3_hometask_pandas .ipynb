```python

file_name= 'Energy Indicators.xls'
column_names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy= pd.read_excel(file_name,skiprows= 17,skipfooter=38,usecols="C:F", names=column_names)
#energy = energy.drop(energy.columns[:2], axis=1)
energy.head(20)
#energy.tail()
```

```python
print(energy.dtypes)
energy['Energy Supply']=energy['Energy Supply'].apply(lambda x: x*1000000 if type(x)==int else (np.NaN if type(x)==str and x=='...' else x))
energy['Energy Supply per Capita']=energy['Energy Supply per Capita'].apply(lambda x: np.NaN if type(x)==str and x=='...' else x)
print(energy.dtypes)
energy.head()
```

```python
energy['Country']=energy['Country'].apply(lambda x: np.NaN if type(x)==str and x=='...' else x)
```

```python
country_dict={'Republic of Korea': 'South Korea',
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'China, Hong Kong Special Administrative Region': 'Hong Kong',    
    'Korea, Rep.': 'South Korea',
    'Iran, Islamic Rep.': 'Iran',
    'Hong Kong SAR, China': 'Hong Kong',
              'Russian Federation':'Mordovia',
              
}

# removing numbers
energy['Country']=energy['Country'].apply(lambda x: ''.join([char for char in x if not char.isdigit()]))
# removing everything in ()
energy['Country']=energy['Country'].apply(lambda x: x[:(x.find('(')-1)] if x.find('(')!=-1 else x)
# changing names
energy['Country']=energy['Country'].apply(lambda x: country_dict[x] if x in country_dict  else x)
#testing changes
energy['Country'].unique()

```

```python
def answer_one ():
    ScimEn_top_15 = ScimEn[ScimEn['Rank'] <= 15]
    GDP_10_years = GDP[['Country Name'] + [str(year) for year in range(2006, 2016)]]
    merged_df = pd.merge(energy, GDP_10_years, how='inner', left_on='Country', right_on='Country Name')
    merged_df = pd.merge(ScimEn_top_15, merged_df, how='inner', left_on='Country', right_on='Country')
    merged_df=merged_df.set_index('Country')
    Column_list=['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    #columns_to_keep = [col for col in Column_list if col in merged_df.columns]
    new_df=merged_df[Column_list]
    return new_df
answer_one ()
```

## Answer the following questions in the context of only the top 15 countries by Scimagojr Rank (aka the DataFrame returned by `answer_one()`)

### Question 2
What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)

*This function should return a Series named `avgGDP` with 15 countries and their average GDP sorted in descending order.*

```python
def answer_two():
    Top15 = answer_one()
    Top15['avgGDP'] = Top15[[str(year) for year in range(2006, 2016)]].mean(axis=1).sort_values(ascending=False)
    avgGDP=pd.Series(Top15['avgGDP'])
    return avgGDP
answer_two()
```

### Question 3
By how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP?

*This function should return a single number.*

```python
def answer_three():
    Top15 = answer_one()
    six_index = answer_two().index[5]
    #print(Top15.loc[six_index,'2006':'2015'])
    GDP_change=Top15['2015'][six_index]-Top15['2006'][six_index]
    return GDP_change
answer_three()
```

### Question 4

Create a new column that is the ratio of Self-Citations to Total Citations. 
What is the maximum value for this new column, and what country has the highest ratio?

*This function should return a tuple with the name of the country and the ratio.*

```python
def answer_four():
    Top15 = answer_one()
    Top15['Ratio_Citation']=Top15['Self-citations']/Top15['Citations']
    result = (Top15['Ratio_Citation'].idxmax(), Top15['Ratio_Citation'].max())
    return result
answer_four()
```

### Question 5

Create a column that estimates the population using Energy Supply and Energy Supply per capita. 
What is the third most populous country according to this estimate?

*This function should return a single string value.*

```python
def answer_five():
    Top15 = answer_one()
    Top15['Estimated_population']=Top15['Energy Supply']/Top15['Energy Supply per Capita']
    result=Top15['Estimated_population'].sort_values(ascending=False).index[2]
    return result
answer_five()
```

### Question 6
Create a column that estimates the number of citable documents per person. 
What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the `.corr()` method, (Pearson's correlation).

*This function should return a single number.*


```python
def answer_six():
    Top15 = answer_one()
    Top15['Estimated_population']=Top15['Energy Supply']/Top15['Energy Supply per Capita']
    Top15['Citable Documents per capita'] = Top15['Citations'] / Top15['Estimated_population']
    #print(Top15.columns)
    correlation = Top15['Citable Documents per capita'].corr(Top15['Energy Supply per Capita'])
    
    return correlation
answer_six()
```

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Mordovia':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
ContinentDict['Brazil']
```

```python
def answer_seven():
    Top15 = answer_one()
    Top15['Estimated_population']=Top15['Energy Supply']/Top15['Energy Supply per Capita']
    Top15['Continent']=Top15.index.map(ContinentDict)
    Cont_Count=Top15.groupby('Continent').count().Rank
    Cont_sum = Top15.groupby('Continent').Estimated_population.sum()
    Cont_mean=Top15.groupby('Continent').Estimated_population.mean()
    Cont_std=Top15.groupby('Continent').Estimated_population.std()
    
    Continent_df=pd.concat([Cont_Count, Cont_sum, Cont_mean,Cont_std], axis=1)
    Continent_df.columns=['Country_Count','Continent_population','Country_avg_population','Country_std_population']
    return Continent_df
answer_seven()
```
